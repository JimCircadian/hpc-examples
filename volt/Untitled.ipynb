```python
import xarray as xr
import datetime as dt

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.dpi'] = 300
```

```python
data = []
with open("data.txt", "r") as fh:
    data = [f for f in fh.read().split("\n") if len(f)]
    
```

```python
power = [(dt.datetime.strptime(line[0:19], "%d-%m-%Y %H:%M:%S"), float(line.split(" ")[-1])) for line in data]
power
```

```python
times = list(zip(*power))[0]
volts = list(zip(*power))[1]
```

```python
fig = plt.figure(figsize=(36, 24))
_, ax = plt.subplots()
ax.scatter(times, volts, s=1)
ax.set_xlabel("Time")
ax.set_ylabel("Supply (V)")
plt.xticks(rotation=90)
ax.set_title("C2 LRAD Supply")
```
